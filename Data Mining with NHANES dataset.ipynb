We now comnine all the data together with join function with SEQN (Respondent Sequence Number)

```python
# Create a new dataset for correlation, displayed via create_report
Income_n2= Income_n.drop(["INDFMMPI", "INDFMMPC"], axis= 1)
Occupation_n2= Occupation_n.drop(["OCQ180", "OCD150", "OCQ210", "OCQ670", "OCD383"], axis= 1)
Hospital_Utilization_n2= Hospital_Utilization_n.drop(["HUQ051", "HUQ071", "HUD062", "HUQ010"], axis= 1)
Immunization_n2= Immunization_n.drop(["IMQ081A", "IMQ081B", "IMQ081C", "IMQ081D", "IMQ090", "IMQ100"], axis= 1)
Hepatitis_n2= Hepatitis_n.copy()
Mental_Health_n2= Mental_Health_n.copy()
Sleep_Disorder_n2= Sleep_Disorder_n.drop(["SLQ300", "SLQ310", "SLD012", "SLQ320", "SLQ330", "SLD013"], axis= 1)
BP_n2= BP_n.drop("BPD035", axis= 1)
```

```python
# We now merging the dataset using outer join to return all record smaple
dfn= Income_n2.merge(Occupation_n2, how= "outer", on= "SEQN")
dfn1= dfn.merge(Hospital_Utilization_n2, how= "outer", on= "SEQN")
dfn2= dfn1.merge(Immunization_n2, how= "outer", on= "SEQN")
dfn3= dfn2.merge(Hepatitis_n2, how= "outer", on= "SEQN")
dfn4= dfn3.merge(Mental_Health_n2, how= "outer", on= "SEQN")
dfn5= dfn4.merge(Sleep_Disorder_n2, how= "outer", on= "SEQN")
dfn6= dfn5.merge(BP_n2, how= "outer", on= "SEQN")
```

**Step 2**: Scale the data and implement a K-Means to determine the clusters


```python
# Now, perform standardization step
# Make a copy of dfn9
dfn10= dfn9.copy()

# Scale the data vis standardization
from sklearn.preprocessing import StandardScaler
dfn10_scaled= StandardScaler().fit_transform(dfn10)
```

```python
# Applying K-Means
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer

model_elb= KMeans()
visualizer_elb= KElbowVisualizer(model_elb, k=(1,22)).fit(dfn10_scaled)
visualizer_elb.show()
```

Observation:
- Instead of using normal way of generating K-Means which rely on subjective persective in deciding and interpreting the numbers of classes within the dataset which high likely to be misinterpretating. We are here creating an elbow plot using KElbowVisualizer from yellowbrick.

- Based on what we can see from the chart generated by using yellowbrick, we can now say that the dataset consist of 6 different classes.

```python
# Applying intercluster distance map
from yellowbrick.cluster import InterclusterDistance

model_id=KMeans(6)
visualizer_id = InterclusterDistance(model_id)
visualizer_id.fit(dfn10_scaled)
visualizer_id.show()
```

Observation:
- Intercluster distance maps display an embedding of the clusters in 2-dimensions with the distance to others centers preserved. An intercluster distance map is often interpreted as below: The closer to the centers, the closer they are in the original feature space. he clusters are sized according to a scoring metric. By default, they are sized by membership, e.g. the number of instances that belong to each center. This gives a sense of the relative importance of clusters. 

- From what we can see, cluster 1 and cluster 3 are very close to original feature space.

Aside from the scaled/ standardized data, let repeat the same steps for unscaled data and explain the differences.

**Step 5**: Implementing Stacked Feature Importance

```python
# Implementing Stacked Feature Importance
from yellowbrick.model_selection import FeatureImportances
from sklearn.linear_model import LogisticRegression

y3= dfn10[["BPQ020"]]
x3= dfn10.drop(["BPQ020"], axis= 1)

model = LogisticRegression(multi_class= "auto", solver= "liblinear")
viz = FeatureImportances(model, stack= True, relative= False)
viz.fit(x3, y3)
viz.show()
plt.show()
```

Ownership: Never assume customer/agree the collection of the data, always ask for permission and identify the data will eb collected.

Privacy: De-identifying of dataset is needed when come into analyzing PPI (personally identifiable information). Some example of PPI are: Name, Address, Phone Number, Credit Card Information, Social Security Card, Bank Account Number, and Passport Number. To enhance the privacy of customer/ client, limiting type of data/ data will be collected is ne of the option.
Intention: When come into data collection and the discussion of data usage, the intention is matter to determine whether a business is violating/ breaching data privacy/ misusing the data.
Outcomes: Despite the intention is good, however if the output of data analytics bring harm/ considered unfair, it can thus can be unlawful.
Data privacy and ethics issues are not limited to those stated above but also included misleading reporting, lack of consideration, excessive data collection, misuse of data, confidentiality of information.

In dealing with data privacy and ethics issues, I believe one should manage privacy proactively rather tahn retrospectively, establish and recognising good data ethics practice, keeping the activities user- centric with strong privacy offering and end to end security throughout the lifecycle of a project.
