**Question 1**
**What is an ensemble classifier? Name some of the popular ensemble methods (EM)? and which one you prefer and why?**

Ensemble learning is regarded as machine learning algorithms that interpret wisdom of crowd/ combination of multiple inducers to solve a particular machine learning task. Ensemble learning is popular as it gather different opinions and weigh for a complex solution, and generally it is better than choosing the opinion of an individual (in other words, improve predictive performance).

Independence, Decentralize, Diversity of opinions, Aggregation.

**Advantages:**

(a) By aggregating numbers of learners, it decrease the risk of obatining local minimum.

(b) By taking into consideration of number of models, the search space would be extends; thus, better fit to data space is achieved.

---
**The application of EM & ML challenges:**

(a) EM is applied to mitigate class imbalance(bias towards majority).

(b) EM solve the idea of concept drift (the changing of features and labels).

(c) EM can be used in eliminate curse of dimenisonality.


In data science languages, EM aim to improve generalizability & robustness over a single estimator.

**There are 2 families of Ensemble Methods: AVERAGEING METHOD & BOOSTING METHOD.**


**AVERAGING METHOD**: Bagging methods & Forests of randomized trees (RandomForest & Extra- Trees)

- By averaging the estimate, it **reduces the variance**

**BOOSTING METHOD**: AdaBoost, Gradient Tree Boosting (Classification & Regression)

- To reduce overfitting, bagging method work best with complex model; while boosting method work best with weaker models- shallow decision tree.

- By boosting the estimate, it **reduces the bias**

This will determine which methods to use in the modelling.

**EM can be separated into 2: The dependent framework & The independent framework**

In dependent framework, previous iteration guide te next iteration.

In independent framework, each inducer built independently.

---

**Types of Ensemble Methods?**

(a) Bagging: Building multiple models from different subset (independently) of training dataset.


(b) Boosting: Building sequence models that attempt to correct mistakes of the model. Boosting algorithms make predictions


(c) Voting: Consider the simplest approach in combining predictions from multiple ML algorithms.

(d) Stacking (less often used): This method generally used in reducing biases. The input is stacked together to compute the prediction. the final estimator is trained via cross- validation.


**Comparisons of Ensemble Methods**

- Boosting provides best results with little noise in data.

- Bagging reduces the variance of unstable methods, while boosting reduced both bias & variance of unstable methods, but increase the variance oof stable methods.

- Boosting is much sensitive to criteria of the dataset & may overfittimg noisy dataset.

- Experiment found that random forest (parallel random)& boosting with 1000 trees have the best average rank.

**In the case of performance prediction, there are 4 factors needed to be considered for selecting best EM and there are:**

(a) Suitability to a given setting

(b) Computational cost

(c) Software availability

(d) Usability



**If I would like to select 2 Ensemble Methods, I would likely select the 2 below:**

**(a) Random Forest**

**(b) Stacked Ensemble Method**


---


**Pros of Random Forest**

- It helps reduces overfitting in decision trees while improve accuracy.
- Work well with both continuous & categorical values.
- Powerful in handling high- dimensional dataset

**Pros of Stacked Ensemble Method**

- Stacking methid reduces both bias & variance

The tree constructured in a top- down manner with categorical attributes. Attributes on the top have more impact than the lower in classifcation as they are idnetified using information gain. Therefore, max_depth can be set to avoid the overfitting issues. **Question 5: Create 3 Machine Learning (Supervised Learning) except any ensemble approach for predicting customer responses.**

It is important to know the differences between ML Classification & Regression

- Classification is used in predicting label

- Regression is used in predicting continuous output

Knowing that 'Customer Response' is binary and it is not continuous variable, therefore, Regression Machine Learning is not suitable in this case, and Classifier Machine Learnign will be applied.

```python
# Import evaluation packages
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix, zero_one_loss, cohen_kappa_score
from sklearn.metrics import matthews_corrcoef
from sklearn.metrics import f1_score
```

```python
# Neural Network
import time
from sklearn.neural_network import MLPClassifier
NN= MLPClassifier(activation= 'logistic', solver= 'sgd', hidden_layer_sizes= 100, random_state= 1)

#Predicting the model with training data & compute time spent
start_time= time.time()
NN_fit= NN.fit(X_train,Y_train)
end_time= time.time()
print("Training time:", end_time-start_time) # 0.5048

#Predicting model with Neural Network
start_time= time.time()
NN_train_PredY= NN.predict(X_train)
NN_Pred_Y = NN.predict(X_test)
end_time= time.time()
print("Training time:", end_time-start_time) #0.006969

# Evaluation the model:
# Classification Report
print(classification_report(Y_test, NN_Pred_Y))

# Accuracy Score
print(accuracy_score(Y_test, NN_Pred_Y))

# Confusion Matrix
confusion_matrix(Y_test, NN_Pred_Y)

# Zero One Loss
print(zero_one_loss(Y_test, NN_Pred_Y))

# Cohen Kappa Score
print(cohen_kappa_score(Y_test, NN_Pred_Y))
```

```python
# ACC/MCC/F1
# Train Dataset
NN_train_accuracy= accuracy_score(Y_train, NN_train_PredY)
NN_train_mcc= matthews_corrcoef(Y_train, NN_train_PredY)
NN_train_f1= f1_score(Y_train, NN_train_PredY, average= 'weighted')

# Test Dataset
NN_test_accuracy= accuracy_score(Y_test, NN_Pred_Y)
NN_test_mcc= matthews_corrcoef(Y_test, NN_Pred_Y)
NN_test_f1= f1_score(Y_test, NN_Pred_Y, average= 'weighted')

print('Performance for Training Set')
print('- Accuracy: %s' % NN_train_accuracy)
print('- MCC: %s' % NN_train_mcc)
print('- F1: %s' % NN_train_f1)
print('------------------------------------------')
print('Performance for Testing Set')
print('- Accuracy: %s' % NN_test_accuracy)
print('- MCC: %s' % NN_test_mcc)
print('- F1: %s' % NN_test_f1)
```

```python
# Evaluation Summary
# Evaluate the train dataset
NN_trainYhat= NN.predict(X_train)
print(f'Train_accuracy:{accuracy_score(Y_train, NN_trainYhat)}')

# Evaluate the test dataset
NN_testYhat= NN.predict(X_test)
print(f'Test_accuracy:{accuracy_score(Y_test, NN_testYhat)}')
```

```python
# Visualization
NN_v= confusion_matrix(Y_test, NN_Pred_Y)
NN_visual= pd.DataFrame(NN_v, columns=np.unique(Y_test), index=np.unique(Y_test))
NN_visual.index.name= 'Actual'
NN_visual.columns.name= 'Predicted'
plt.figure(figsize= (10,8))
sns.set(font_scale=1)
sns.heatmap(NN_visual,cmap= 'Greens',annot= True)
plt.show()
```

```python
# Stacked Ensemble Method
from sklearn. ensemble import StackingClassifier
from sklearn.metrics import matthews_corrcoef
from sklearn.metrics import f1_score
from sklearn.linear_model import LogisticRegression

LR= LogisticRegression()

estimators=[
    ('KNN', KNN),
    ('NN', NN),
    ('SVM', SVM)]

# BUild a stack model
Stackmodel= StackingClassifier(estimators= estimators, final_estimator= LR)

# Train the stacked model
start_time= time.time()
Stack_model= Stackmodel.fit(X_train, Y_train)
end_time= time.time()
print("Training time:", end_time-start_time)

# Predicting model using Stackmodel
start_time= time.time()
Stack_Pred_Y= Stackmodel.predict(X_test)
Stack_Pred_YTrain= Stackmodel.predict(X_train)
end_time= time.time()
print("Training time:", end_time-start_time)

# Print accuracy of the train dataset
SM_train_accuracy= accuracy_score(Y_train, Stack_Pred_YTrain)
SM_train_mcc= matthews_corrcoef(Y_train, Stack_Pred_YTrain)
SM_train_f1= f1_score(Y_train, Stack_Pred_YTrain)

# Print accuracy of test dataset
SM_test_accuracy= accuracy_score(Y_test, Stack_Pred_Y)
SM_test_mcc= matthews_corrcoef(Y_test, Stack_Pred_Y)
SM_test_f1= f1_score(Y_test, Stack_Pred_Y)

#Confusion matrix
confusion_matrix(Y_test,Stack_Pred_Y)

#Evaluating the model
#Classification report
print(classification_report(Y_test,Stack_Pred_Y))

#Zero one loss
print(zero_one_loss(Y_test,Stack_Pred_Y))

#Kappa
print(cohen_kappa_score(Y_test,Stack_Pred_Y))


print('Performance for Training Set')
print('- Accuracy: %s' % SM_train_accuracy)
print('- MCC: %s' % SM_train_mcc)
print('- F1: %s' % SM_train_f1)
print('------------------------------------------')
print('- Accuracy: %s' % SM_test_accuracy)
print('- MCC: %s' % SM_test_mcc)
print('- F1: %s' % SM_test_f1)
```
